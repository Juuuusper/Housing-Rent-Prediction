```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_data = pd.read_csv('D:\pythonaa/train1.csv' ,encoding='utf-8')
```

```python
print(train_data.isnull().sum())
```

```python
train_data.head(10)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(211)
plt.title('Price Distribution')
sns.distplot(train_data['price'])#价格的数量分布曲线
plt.subplot(212)
plt.scatter(range(train_data.shape[0]),np.sort(train_data['price'].values))
plt.show()
```

```python
train_data.describe()
```

```python
train_data.hist(figsize=(20,15),bins=50,grid=False)
plt.show()
```

```python
continuous_cols=['name','number','allfloor','housesize','postion','sub_postion','distance']
for col in continuous_cols:
    sns.jointplot(x=col,y='price',data=train_data,alpha=0.3,size=4)
```

```python
plt.figure(figsize=(12,6))
train_data.corr()['price'][continuous_cols].sort_values(ascending=False).plot(
    'barh',figsize=(12,6),title='Variable Correlation with Price'
)
```

```python
plt.figure(figsize=(12,10))
plt.subplot(211)
sns.boxplot(y='floor',x='price',data=train_data,orient='h')
plt.subplot(212)
sns.boxplot(y='bedroom',x='price',data=train_data,orient='h')
```

```python
plt.figure(figsize=(12,10))
plt.subplot(211)
sns.boxplot(y='hall',x='price',data=train_data,orient='h')
plt.subplot(212)
sns.boxplot(y='guard',x='price',data=train_data,orient='h')
```

```python
plt.figure(figsize=(12,10))
plt.subplot(211)
sns.boxplot(y='area',x='price',data=train_data,orient='h')
plt.subplot(212)
sns.boxplot(y='sub_num',x='price',data=train_data,orient='h')
```

```python
from scipy.stats import spearmanr

r,p=spearmanr(train_data['floor'],train_data['price'])
print('floor和price的相关系数为 %s,其中p=%s'%(r,p))
r,p=spearmanr(train_data['bedroom'],train_data['price'])
print('bedroom和price的相关系数为 %s,其中p=%s'%(r,p))
r,p=spearmanr(train_data['hall'],train_data['price'])
print('hall和price的相关系数为 %s,其中p=%s'%(r,p))
r,p=spearmanr(train_data['guard'],train_data['price'])
print('guard和price的相关系数为 %s,其中p=%s'%(r,p))
r,p=spearmanr(train_data['area'],train_data['price'])
print('area和price的相关系数为 %s,其中p=%s'%(r,p))
r,p=spearmanr(train_data['sub_num'],train_data['price'])
print('sub_num和price的相关系数为 %s,其中p=%s'%(r,p))
```
